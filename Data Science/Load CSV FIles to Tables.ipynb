### Imports the required functions

```python
from pyspark.sql.types import LongType, IntegerType, DateType, StringType
from pyspark.sql.functions import to_date
```

### Reads the data from the ***DimDates.csv*** file and store the data in the ***dfDates*** spark data frame and casts each column to the correct data type

```python
dfDates = spark.read.format("csv").option("header","true").load("Files/DimDates.csv")
```

```python
dfDates = dfDates.withColumn('DateKey', dfDates.DateKey.cast(IntegerType())) \
  .withColumn('Date', to_date(dfDates.Date, 'M/d/yyyy').cast(DateType())) \
  .withColumn('MonthYr', dfDates.MonthYr.cast(StringType())) \
  .withColumn('MonthYrKey', dfDates.MonthYrKey.cast(IntegerType())) \
  .withColumn('Year', dfDates.Year.cast(IntegerType())) \
  .withColumn('QuarterYear', dfDates.QuarterYear.cast(StringType())) \
  .withColumn('QuarterYearKey', dfDates.QuarterYearKey.cast(IntegerType())) \
  .withColumn('Quarter', dfDates.Quarter.cast(StringType())) \
  .withColumn('QuarterKey', dfDates.QuarterKey.cast(IntegerType())) \
  .withColumn('Month', dfDates.Month.cast(StringType())) \
  .withColumn('MonthKey', dfDates.MonthKey.cast(IntegerType())) \
  .withColumn('Weekday', dfDates.Weekday.cast(StringType())) \
  .withColumn('WeekdayKey', dfDates.WeekdayKey.cast(IntegerType())) \
  .withColumn('Week', to_date(dfDates.Week, 'M/d/yyyy').cast(DateType()))
```

### Creates a table named ***dates*** based on the contents in the ***dfDates*** data frame

```python
%%sql

DROP TABLE IF EXISTS dates
```

```python
dfDates.write.mode("overwrite").format("delta").save("Tables/dates")
```
